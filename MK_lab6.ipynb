```python
import numpy as np
import matplotlib.pyplot as plt

# DEFINICJE FUNKCJI

def sigmoid(z):
    """
    Oblicza wartość funkcji sigmoidalnej.
    Wzór: 1 / (1 + e^-z)
    """
    return 1 / (1 + np.exp(-z))

def sigmoid_derivative(z):
    """
    Oblicza pochodną (gradient) funkcji sigmoidalnej.
    Wzór: sigmoid(z) * (1 - sigmoid(z))
    """
    s = sigmoid(z)
    return s * (1 - s)

# GENEROWANIE DANYCH

# Zakres argumentów (z) od -10 do 10
z = np.linspace(-10, 10, 400)

# Obliczenie wartości funkcji i pochodnej
y = sigmoid(z)
y_grad = sigmoid_derivative(z)

# WIZUALIZACJA

plt.figure(figsize=(10, 6))

# Wykres funkcji sigmoidalnej
plt.plot(z, y, label=r'Funkcja aktywacji $\sigma(z)$', color='blue', linewidth=2)

# Wykres pochodnej
plt.plot(z, y_grad, label=r'Gradient $\sigma\'(z)$', color='red', linestyle='--', linewidth=2)

# Dodatki do wykresu
plt.title("Funkcja Sigmoidalna i jej pochodna", fontsize=14)
plt.xlabel("Wejście (z)", fontsize=12)
plt.ylabel("Wartość", fontsize=12)
plt.grid(True, alpha=0.6)
plt.legend(fontsize=12)
plt.axhline(0, color='black', linewidth=0.8) # Oś X
plt.axvline(0, color='black', linewidth=0.8) # Oś Y

# Zaznaczenie zakresu wartości
plt.ylim(-0.1, 1.1)

plt.show()
```
